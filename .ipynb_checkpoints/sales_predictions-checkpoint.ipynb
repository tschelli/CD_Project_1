```python
#Loop through the index of the Dataframe
for ind in eda_df.index:
    #Create a variable to hold the Item_Identifier of the current index
    item_id = eda_df.loc[ind,'Item_Identifier']

    #Check if the location of the Item_Weight at the index is null
    #Create a filter to filter out only 'Item_Identifier' values equivalent to the item_id
    #Replace the null value with the .mean() of the 'Item_Weight' for the current 'Item_Identifier'. Values across 'Item_Identifier's are consistent.
    if eda_df.isnull().loc[ind, 'Item_Weight']:
        item_filter = eda_df['Item_Identifier'] == item_id
        eda_df.loc[ind, 'Item_Weight'] = eda_df[item_filter]['Item_Weight'].mean()

        #Additional check for the case of the null 'Item_Weight' being the only value for the 'Item_Identifier'
        #Create a new filter to filter out only the 'Item_Type' values of the current index.
        #Replace the null values with the .mean() of the 'Item_Weight' for the item's 'Item_Type'
        if math.isnan(eda_df[item_filter]['Item_Weight'].mean()):
            type_filter = eda_df['Item_Type'] == eda_df.loc[ind, 'Item_Type']
            eda_df.loc[ind, 'Item_Weight'] = eda_df[type_filter]['Item_Weight'].median()
```

- Unique `Item_Identifier`'s have consistent `Item_Weight`'s.
    - The logic that was run updates null `Item_Weight` values with it's relevant `Item_Identifier`'s `Item_Weight` value.
- In the case that the `Item_Identifier` only had null `Item_Weight` entries, the `Item_Weight` will be filled using the median `Item_Weight` of it's `Item_Type`.

### Create Functions

```python
score_compare_df = pd.DataFrame({'Metrics':['MAE', 'MSE', 'RMSE', 'R2']})
score_compare_df.set_index('Metrics', inplace = True)

def eval_models(true_train, pred_train, true_test, pred_test, model):
#Takes true and predicted values and prints MAE, MSE, RMSE and R2 for both the train and test data
    mae_train = round(mean_absolute_error(true_train, pred_train), 4)
    mse_train = round(mean_squared_error(true_train, pred_train), 4)
    rmse_train = round(np.sqrt(mse_train), 4)
    r2_train = round(r2_score(true_train, pred_train), 4)

    mae_test = round(mean_absolute_error(true_test, pred_test), 4)
    mse_test = round(mean_squared_error(true_test, pred_test), 4)
    rmse_test = round(np.sqrt(mse_test), 4)
    r2_test = round(r2_score(true_test, pred_test), 4)

    train_name = (f'{model} Training Data')
    test_name = (f'{model} Testing Data')
    train_list = [mae_train, mse_train, rmse_train, r2_train]
    test_list = [mae_test, mse_test, rmse_test, r2_test]
    score_compare_df[train_name] = train_list
    score_compare_df[test_name] = test_list

    print(f'Training Scores for {model}:\nMAE:    {mae_train}\n MSE:    {mse_train}\n RMSE: {rmse_train}\n R^2:     {r2_train}\n')
    print(f'Testing Scores for {model}:\nMAE:    {mae_test}\n MSE:    {mse_test}\n RMSE: {rmse_test}\n R^2:     {r2_test}\n')
```

```python
pd.set_option('display.float_format', lambda x: '%.4f' % x)
```

```python
def score_plot(test_scores_df, train_scores_df, max_depth, scores_df, steps):
    plt.plot(test_scores_df)
    plt.plot(train_scores_df)
    plt.xticks(np.arange(0, max_depth+2, step=steps))
    plt.xlabel('max_depth')
    plt.ylabel('R2')
    plt.show()

    sorted_scores = scores_df.sort_values(by='Test Score', ascending=False)
    display(sorted_scores.head())
    score_plot.top_score = sorted_scores.index[0]
```

### Impute Missing Item_Weights

```python
#Loop through the index of the Dataframe
for ind in X_train.index:
    #Create a variable to hold the Item_Identifier of the current index
    item_id = X_train.loc[ind,'Item_Identifier']

    #Check if the location of the Item_Weight at the index is null
    #Create a filter to filter out only 'Item_Identifier' values equivalent to the item_id
    #Replace the null value with the .mean() of the 'Item_Weight' for the current 'Item_Identifier'. Values across 'Item_Identifier's are consistent.
    if X_train.isnull().loc[ind, 'Item_Weight']:
        item_filter = X_train['Item_Identifier'] == item_id
        X_train.loc[ind, 'Item_Weight'] = X_train[item_filter]['Item_Weight'].mean()

        #Additional check for the case of the null 'Item_Weight' being the only value for the 'Item_Identifier'
        #Create a new filter to filter out only the 'Item_Type' values of the current index.
        #Replace the null values with the .mean() of the 'Item_Weight' for the item's 'Item_Type'
        if math.isnan(X_train[item_filter]['Item_Weight'].mean()):
            type_filter = X_train['Item_Type'] == X_train.loc[ind, 'Item_Type']
            X_train.loc[ind, 'Item_Weight'] = X_train[type_filter]['Item_Weight'].median()
```

```python
# Column selectors
num_selector = make_column_selector(dtype_include='number')

# Split ordinal columns out
ord_cols = ['Outlet_Size', 'Outlet_Location_Type']
cat_cols = ml_df.select_dtypes('object').columns.difference(['Outlet_Size', 'Outlet_Location_Type', 'Item_Identifier']).to_list()
cat_cols
```

### Instantiate Transformers

```python
# Simple Imputer
freq_imputer = SimpleImputer(strategy = 'most_frequent')
# Scaler
scaler = StandardScaler()
# One Hot Encoder
ohe = OneHotEncoder(handle_unknown='ignore', sparse=False)
```

### Ordinal Encoder

```python
# Create a list of ordinal labels
out_size_labels = ['Small', 'Medium', 'Large', 'Unknown']
out_loc_type = ['Tier 1', 'Tier 2', 'Tier 3']

# Combine the ordered list in the order that the columns appear
ordered_labels = [out_size_labels, out_loc_type]

# Instantiate OrdinalEncoder
ordinal = OrdinalEncoder(categories = ordered_labels)
```

### Instantiate Pipelines

```python
# Pipeline for numerals and categories
ord_pipe = make_pipeline(freq_imputer, ordinal)
cat_pipe = make_pipeline(freq_imputer, ohe)
```

### Instantiate Column Transformer

```python
# Setup the tuples to pair the processors with the make column selectors
num_tuple = (scaler, num_selector)
ord_tuple = (ord_pipe, ord_cols)
cat_tuple = (cat_pipe, cat_cols)

# Make column transformer
preprocessor = make_column_transformer(num_tuple, cat_tuple, ord_tuple, remainder = 'passthrough')
```

#### Tune the Decision Tree

```python
#List of values to try for max_depth
depths = list(range(2, dec_tree.get_depth())) 
#Dataframe to store the score for each value of max_depth
scores_dt = pd.DataFrame(index=depths, columns=['Test Score','Train Score'])
for depth in depths:
    dec_tree = DecisionTreeRegressor(max_depth=depth, random_state=42)
    dec_tree_pipe = make_pipeline(preprocessor, dec_tree)
    dec_tree_pipe.fit(X_train, y_train)
    train_score = dec_tree_pipe.score(X_train, y_train)
    test_score = dec_tree_pipe.score(X_test, y_test)
    scores_dt.loc[depth, 'Train Score'] = train_score
    scores_dt.loc[depth, 'Test Score'] = test_score
```

### Random Forests Model

```python
#Instantiate the model
rf = RandomForestRegressor(random_state = 42)
#Make a pipeline
rf_pipe = make_pipeline(preprocessor, rf)
```

```python
#Fit and evaluate the model
rf_pipe.fit(X_train, y_train)

eval_models(y_train, rf_pipe.predict(X_train), y_test, rf_pipe.predict(X_test), 'Random Forests')
```

#### Tune the Random Forests Model Depth

```python
#Determine the depth of each tree in the random forest
est_depths = [estimator.get_depth() for estimator in rf.estimators_]
```

```python
#Range from 1 to the max est_depth
depths = range(1, max(est_depths))
#Dataframe to store the scores
scores_rf = pd.DataFrame(columns=['Train Score','Test Score'], index=depths)
#Iterate through the values to find the best depth
for depth in depths:
    rf = RandomForestRegressor(max_depth = depth, random_state = 42)
    rf_pipe = make_pipeline(preprocessor, rf)
    rf_pipe.fit(X_train, y_train)
    train_score = rf_pipe.score(X_train, y_train)
    test_score = rf_pipe.score(X_test, y_test)
    scores_rf.loc[depth, 'Train Score'] = train_score
    scores_rf.loc[depth, 'Test Score'] = test_score
```

```python
# Plot the scores and display sorted scores
score_plot(scores_rf['Test Score'], scores_rf['Train Score'], max(est_depths), scores_rf, 2)
```

#### Tune the Random Forests Estimators

```python
#List of estimator values
n_ests = range(20,320, 20)
#Dataframe to store the scores
scores_rf_est = pd.DataFrame(index=n_ests, columns=['Test Score', 'Train Score'])
#Iterate through the values to find the best number of estimators with optimized depth
for n in n_ests:
    rf = RandomForestRegressor(max_depth=score_plot.top_score, n_estimators=n, random_state = 42)
    rf_pipe = make_pipeline(preprocessor, rf)
    rf_pipe.fit(X_train, y_train)
    train_score = rf_pipe.score(X_train, y_train)
    test_score = rf_pipe.score(X_test, y_test)
    scores_rf_est.loc[n, 'Train Score'] = train_score
    scores_rf_est.loc[n, 'Test Score'] = test_score
```

```python
# Plot the scores and display sorted scores
score_plot(scores_rf_est['Test Score'], scores_rf_est['Train Score'], max(n_ests), scores_rf_est, 20)
```
